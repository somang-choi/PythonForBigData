```python
population_seoul = pandas.read_excel(
    '../data/population_in_Seoul.xls',
    header = 2,
    parse_cols = 'B, D, G, J, N',
    encoding='utf-8'
)
```

```python
population_seoul.head()
```

```python
population_seoul.rename(
    columns={
        population_seoul.columns[0]: '구별',
        population_seoul.columns[1]: '인구수',
        population_seoul.columns[2]: '한국인',
        population_seoul.columns[3]: '외국인',
        population_seoul.columns[4]: '고령자',
    },
    inplace=True
)
```

```python
population_seoul.head()
```

```python
CCTV_Seoul.tail()
```

```python
CCTV_Seoul.sort_values(by='소계', ascending=False).head()
```

```python
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2014년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2016년']) / CCTV_Seoul['2013년도 이전'] * 100
```

```python
CCTV_Seoul.sort_values(by='최근증가율', ascending=False).head()
```

```python
population_seoul[population_seoul['구별'].isnull()]
```

```python
population_seoul.drop([26], inplace=True)
```

```python
population_seoul.tail()
```

```python
population_seoul['외국인비율'] = (population_seoul['외국인'] / population_seoul['인구수']) * 100
population_seoul['고령자비율'] = (population_seoul['고령자'] / population_seoul['인구수']) * 100
```

```python
population_seoul.head()
```

```python
population_seoul.sort_values(by='외국인비율', ascending=False).head()
```

```python
population_seoul.sort_values(by='고령자비율', ascending=False).head()
```

```python
population_seoul.sort_values(by='고령자', ascending=False).head()
```

```python
cctv_with_population = pandas.merge(CCTV_Seoul, population_seoul, on='구별')
cctv_with_population.head()
```

```python
del cctv_with_population['2013년도 이전']
del cctv_with_population['2014년']
del cctv_with_population['2015년']
del cctv_with_population['2016년']
```

```python
# matplotlib 이 기본으로 가진 폰트는 한글을 지원하지 않음
# matplotlib 의 폰트 변경하기
# 강사가 공유해 주는 malgun.ttf (맑은 고딕) 폰트를
# C:\fonts 디렉토리 생성 후 복사
# C:\fonts\malgun.ttf

from matplotlib import pyplot
%matplotlib inline

import platform

from matplotlib import font_manager, rc
pyplot.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')
```
