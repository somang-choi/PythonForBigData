```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "key": list("bbacaab"),
    "data1": range(7)
})
df2 = pd.DataFrame({
    "key": list("abd"),
    "data2": range(3)
})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='key') #df2에 없는 key인 c에 대한 결과는 없음
```

```python
pd.merge(df1, df2, on='key', how='left') # df2에 없는 key에 대해 NaN이 대입됨
```

```python
pd.merge(df1, df2, on='key', how='outer')
```

```python
df1 = pd.DataFrame({
    "key": list("bbacab"),
    "data1": range(6)
})
df2 = pd.DataFrame({
    "key": list("ababd"),
    "data2": range(5)
})
# df1에도 동일 key에 대한 값이 복수 개 있고, df2에도 동일 key에 대한 값이 복수 개 있는 data
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='key', how='inner') # 데카르트 곱
```

```python
pd.merge(df1, df2, on='key', how='outer') 
```
