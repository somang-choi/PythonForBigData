```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as pyplot
```

```python
# 모든 module이 __version__ 속성을 가짐
# 버전에 따라 그래프 모양이 다를 수 있음
matplotlib.__version__ 
pd.__version__
```

```python
sr = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
```

```python
sr
```

```python
sr.plot()
pyplot.savefig('figure1.png')
```

```python
df = pd.DataFrame(
    np.random.randn(10, 4).cumsum(axis=0),
    columns=["A", "B", "C", "D"],
    index=np.arange(0, 100, 10)
)
```

```python
df.plot()
pyplot.savefig('figure2.png')
```

```python
df2 = pd.DataFrame(
    np.random.rand(6, 4),
    index=["one", "two", "three", "four", "five", "six"],
    columns=pd.Index(["A", "B", "C", "D"], name="Number")
)
```

```python
df2
```

```python
df2.plot(kind='bar')
df2.plot(kind='barh')
# legend는 legend으로 인해 그래프를 보기에 방해되지 않는 최적의 위치에 표시되므로 그때 그때 다름
```

```python
df2.plot(kind='barh', stacked=True)
# stacked=True 인자를 넣으면 하나의 index에 대응되는 각 열 값의 상대적 비율을 확인하기 좋음
```

```python
x1 = np.random.normal(1, 1, size=(100, 1))
x2 = np.random.normal(-2, 4, size=(100, 1))
```

```python
X = np.concatenate((x1, x2), axis=1)
```

```python
X
```

```python
df3 = pd.DataFrame(X, columns=['x1', 'x2'])
```

```python
df3
```

```python
pyplot.scatter(df3['x1'], df3['x2'])
pyplot.show() # 텍스트 결과를 보여주지 않고 그래프 결과만 보여줌
```
