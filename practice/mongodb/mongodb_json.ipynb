```python
import pandas, pymongo, json
from pymongo import MongoClient
```

```python
url = 'localhost'
port = 27017
client = MongoClient(url, port)
```

```python
db = client.jsondb # 있으면 연결, 없으면 생성한다
```

```python
collection = db.test
```

```python
djson = open('bitcoin_sample.json').read()
data = json.loads(djson) # json.dump
```

```python
collection.insert_many(data)
```

```python
for item in collection.find():
    print(item['name'])
```
